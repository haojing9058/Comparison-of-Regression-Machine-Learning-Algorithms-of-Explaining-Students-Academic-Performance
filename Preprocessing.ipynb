### Preprocessing student-mat.csv

```python
import pandas as pd
```

```python
#load data into rawdata
dfm = pd.read_csv('student-mat.csv',sep=';')
```

```python
dfm.head()
```

```python
dfm.columns
```

```python
dfm.dtypes
```

```python
dfm.select_dtypes(include=['object'])
```

```python
#Dummy code categorical variables.
dfm = pd.get_dummies(dfm)
dfm.head()
```

```python
dfm.dtypes
```

```python
#center and scale varialbes
from sklearn import preprocessing
import numpy as np
```

```python
X = np.array(dfm)
X
```

```python
X_scaled = preprocessing.scale(X)
X_scaled
```

```python
X_scaled.mean(axis=0)
```

```python
X_scaled.std(axis=0)
```

```python
X_centered = preprocessing.scale(X,with_mean=True, with_std=False)
X_centered
```

```python
X_centered.mean(axis=0)
```
